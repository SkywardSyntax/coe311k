```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import math

def factorial(n):
    if (n == 1):
        return 1
    
    return(n * factorial(n-1))


def tSeriesSin(numTerms, inputX):
    if (numTerms < 1):
        sys.exit("Invalid Number of Terms Provided (n<1)")
        
    if (numTerms == 1):
        return inputX
    
    n = 2*(numTerms)-1
    
    nFact = (factorial(n))
        
    return (((-1)**(numTerms-1) * (inputX**((n))/nFact) + tSeriesSin(numTerms - 1, inputX)))

x = np.arange(0,10, .1)
maxApproxIterations = 4

approx_answers = np.empty(maxApproxIterations, dtype=object)
exact_answers = np.sin(x)

for i in range (maxApproxIterations):
      plt.subplot(2,2,i+1)
      approx_answers[i] = tSeriesSin(i+1, x)
      plt.plot(x, abs(approx_answers[i] - exact_answers))
      plt.title(f'Iteration {i+1}')
 

plt.tight_layout()
plt.show()
```

DISCUSSION QUESTIONS:

As you include more terms the error around small values of x tends to be really small.

Furthermore, the more iterations we do, the farther we seem to be able to progress from x = 0 while still having a good estimate of sin(x). 

However, as we add more terms, for large x values (closer to 10), we are getting increasingly worse estimations of the sin function (error seems to be diverging faster as we add more terms)
