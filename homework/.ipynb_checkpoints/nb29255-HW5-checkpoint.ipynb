```python
import numpy as np

def inv(A):
    return np.linalg.inv(A)

points = np.array([[0,1],[1,5],[5,1]])

A = np.zeros((len(points[0]), len(points))).T


B=np.zeros(points.shape[0]).T
           
for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        if (j == 0):
            A[i][j] = 1
        else:
            A[i][j] = points[i][j-1]
            B[i]=points[i][j]

print(inv(A.T@A)@(A.T)@B)

#first element would be c1 and second element would be c2 where c1 is y-intercept and x2 is slope




```
