```python
import random
import numpy as np
import matplotlib.pyplot as plt

def is_in_circle(x, y, radius=1.0):
    return (x*x + y*y <= radius*radius)

def monte_carlo_pi(num_samples):
    num_samples = int(num_samples)
    
    points_in_circle = 0
    for i in range(num_samples):
        x = random.random()
        y = random.random()
        if is_in_circle(x, y):
            points_in_circle += 1
    
    return 4 * points_in_circle / num_samples

sample_sizes = np.arange(10, 20000, 20)
pi_ests = np.zeros(sample_sizes.size, dtype=np.float64)

for i in range(pi_ests.size):
    pi_ests[i] = monte_carlo_pi(sample_sizes[i])

pi_errors = np.abs(np.pi - pi_ests)

plt.plot(sample_sizes, pi_errors)
plt.show()
```
