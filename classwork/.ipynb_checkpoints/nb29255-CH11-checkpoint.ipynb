```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

# Define known data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

# Create a cubic spline interpolation of the data
cs = CubicSpline(x, y, bc_type='natural')

# Create a polynomial fit of the data (degree 5 to pass through all 6 points)
coeffs = np.polyfit(x, y, deg=3)
poly = np.poly1d(coeffs)

# Generate new x values for a smooth plot
x_interp = np.linspace(0, 5, 100)
y_spline = cs(x_interp)
y_poly = poly(x_interp)

# Plot the original data, cubic spline, and polynomial fit
plt.plot(x, y, 'bo', label="Data points")
plt.plot(x_interp, y_spline, label="Cubic Spline Interpolation")
plt.plot(x_interp, y_poly, label="Polynomial Fit (degree 5)")
plt.title("Cubic Spline vs Polynomial Fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

We're actually hitting the points with out cubic spline interpolation, instead of just minimizing the residuals with a fixed amount of precision (relating to the order of the polynomial interpolation).
