```python
import numpy as np
import matplotlib.pyplot as plt


# Data points
X = np.arange(0,1,0.01)
linEq = (1/2)*X
Y = linEq + np.random.uniform(-1,1,X.size)


# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T

# Solve using least squares
m, c = np.linalg.lstsq(A, linEq, rcond=None)[0]
m1, c1 = np.linalg.lstsq(A, Y, rcond=None)[0]


print(f"Slope: {m}, Intercept: {c}")


plt.scatter(X, Y)
X_line = np.linspace(X.min(), X.max(), 100)
Y_line=m*X_line+c
Y_line1 = m1 * X_line + c1

plt.plot(X_line, Y_line1, color='black') # Line w/ Random Perturbations
plt.plot(X_line, Y_line, color='red') # Line w/o Random Perturbations
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```
