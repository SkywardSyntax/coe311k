```python
import random
import math
import numpy as np

def estimate_pi(num_samples):
    points_inside_circle = 0
    
    for _ in range(num_samples):
        x = random.random()
        y = random.random()
        distance = math.sqrt(x**2 + y**2)
        if distance <= 1:
            points_inside_circle += 1
    
    pi_estimate = 4 * points_inside_circle / num_samples
    return pi_estimate


samples_1000 = 1000
samples_100 = 100

estimated_pi_1000 = estimate_pi(samples_1000)
estimated_pi_100 = estimate_pi(samples_100)


actual_pi = np.pi

abs_error_1000 = abs(actual_pi - estimated_pi_1000)
abs_error_100 = abs(actual_pi - estimated_pi_100)



rel_error_1000 = abs_error_1000 / actual_pi
rel_error_100 = abs_error_100 / actual_pi


print(f"\n1,000 samples:")
print(f"  Estimated π: {estimated_pi_1000}")
print(f"  Absolute Error: {abs_error_1000}")
print(f"  Relative Error: {rel_error_1000}")

print(f"\n100 samples:")
print(f"  Estimated π: {estimated_pi_100}")
print(f"  Absolute Error: {abs_error_100}")
print(f"  Relative Error: {rel_error_100}")
```
