```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(0,2*np.pi,99)
y = np.linspace(0,1,20)
alpha = np.random.rand(1000)
beta = np.random.rand(1000)
size = np.random.rand(1000) * 12.5
color = np.random.rand(1000)
gamma = alpha > 0.5
sigma = alpha
sigma[gamma] = 1
sigma[~gamma] = 0


plt.subplot(2,2,1)
plt.plot(x, np.sin(x))
plt.subplot(2,2,2)
plt.hist(sigma)
plt.subplot(2,2,3)
plt.scatter(alpha,beta,size,color)
plt.colorbar()
plt.subplot(2,2,4)
plt.scatter(y, np.arcsin(y))

plt.show()
```
