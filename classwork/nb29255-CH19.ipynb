```python
import numpy as np
import matplotlib.pyplot as plt

# ODE: dy/dt = -1000y + 3000 - 2000 e^-t,  y(0)=0
# Analytical: y(t) = 3 - 0.998 e^-1000t - 2.002 e^-t


k = 1000.0 


h = np.array([0.0005, 0.001, 0.0015])

t_start = 0.0
t_end = 0.0005
y_0 = 0.0

def f(t, y):
    return -k * y + 3000.0 - 2000.0 * np.exp(-t)

def y_exact(t):
    return 3.0 - 0.998 * np.exp(-1000.0 * t) - 2.002 * np.exp(-t)

# Plot exact solution once
t_fine = np.linspace(t_start, t_end, 5000)
plt.figure(figsize=(10, 5))
plt.plot(t_fine, y_exact(t_fine), 'k', linewidth=2, label='Analytical y(t)')

# Euler Forward for each step size
for step in range(len(h)):
    dt = h[step]
    t_values = np.arange(t_start, t_end + dt, dt)
    y_values = np.zeros(len(t_values))

    # Set initial condition
    y_values[0] = y_0

    # Euler's Method loop
    for i in range(1, len(t_values)):
        t_n = t_values[i - 1]
        y_n = y_values[i - 1]
        y_values[i] = y_n + dt * f(t_n, y_n)

    plt.plot(t_values, y_values, label="Euler h= " + str(dt))
plt.ylim(-10, 10)
plt.xlabel('t')
plt.ylabel('y')

plt.title("Forward Euler vs Analytical: dy/dt = -1000y + 3000 - 2000e^{-t}")
plt.legend()
plt.grid(True)
plt.show()
```
