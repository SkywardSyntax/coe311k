```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import math

def factorial(n):
    if (n == 1):
        return 1
    
    return(n * factorial(n-1))


def tSeriesSin(numTerms, inputX):
    if (numTerms < 1):
        sys.exit("Invalid Number of Terms Provided (n<1)")
        
    if (numTerms == 1):
        return inputX
    
    n = 2*(numTerms)-1
    
    nFact = (factorial(n))
        
    return (((-1)**(numTerms-1) * (inputX**((n))/nFact) + tSeriesSin(numTerms - 1, inputX)))

x = np.arange(-12*np.pi,12*np.pi, 0.1)
plt.plot(x, tSeriesSin(85, x))
plt.plot(x, np.sin(x))
plt.show()
```
