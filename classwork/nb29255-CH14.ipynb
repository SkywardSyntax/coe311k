```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

def f_prime(x):
    return np.cos(x)

    
b = 2*np.pi
a = -2*np.pi
n = 100

x = np.linspace(a, b, n)
h = x[1] - x[0]




# Central Difference Approximation
df_dx  = (f(x + h) - f(x - h)) / (2*h)

# Forward Difference Approximation
### YOUR CODE HERE ###
df_dx2 = (f(x + h) - f(x)) / h

# Backward Difference Approximation
### YOUR CODE HERE ###
df_dx3 = (f(x) - f(x - h)) / h

# Plot the approximation and the true derivative
## Do a canvas of 4 subplots or 4 graphs on one plot comparing cos(x), Forward Difference, Backwards Difference, and Central Difference Methods


plt.subplot(3, 2, 1)
plt.title("Cos(x)")
plt.plot(x, f_prime(x), 'g')


plt.subplot(3,2,2)
plt.title("Cos(x) - Central")
plt.plot(x, df_dx, 'y')


plt.subplot(3,2,5)
plt.title("Cos(x) - Forward")
plt.plot(x, df_dx2, 'b')

plt.subplot(3,2,6)
plt.title("Cos(x) - Backward")
plt.plot(x, df_dx3, 'y')

plt.show()
```
