```python
import numpy as np
import matplotlib.pyplot as plt
import random





#I'm treating least squares as a special case of Polynomial Fit (polynomial of degree 1)

def polyFit(x_in,y,degree):
    if (degree < 0):
        return -1
    x = np.ones(y.size)
    if (degree < 1):
        return (x@y)/x.size
        
    #Constructs the vandermonde matrix
    for i in range(1, degree+1):
        x = np.column_stack((x, x_in**i))
    
    return np.linalg.inv(x.T@x)@(x.T@y) #result will always be in form nx1, where n is number of cols in x
#                        nxmXmxn



x = np.linspace(0,19,20)
gdp_growth = np.array([1.7, 0.1, 2.3, 2.7, 1.7, 5.0, 3.0, 1.5, 3.5, 1.8, 1.3, 2.5, 3.1, -5.1, -28.1, 33.8, 6.3, -1.6, 2.4, 3.3])
#degree N polynomial
#degree N polynomial has N+1 Terms
n=4
degreeNCoeffs = polyFit(x,gdp_growth,n)
degreeNPoly = np.zeros(x.size)


for i in range(0, n+1):
    degreeNPoly = degreeNPoly + degreeNCoeffs[i]*x**i
    
plt.scatter(x, gdp_growth, color='red', zorder=5)
plt.plot(x, degreeNPoly)
plt.plot(x, degreeNPoly)
```
